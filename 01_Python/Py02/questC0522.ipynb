###  Q1. 클로저를 사용하여 주어진 숫자 리스트에서 최댓값과 최솟값을 찾는
### 함수를 작성하세요.

조건:
 ●주어진 숫자 리스트 numbers에 대해 find_min_max 함수를 호출하면 
find_min과 find_max 함수를 얻을 수 있습니다.
 ● find_min() 함수를 호출하면 최솟값을, find_max() 함수를 호출하면 
최댓값을 반환합니다.
 ●클로저를 사용하여 min_value와 max_value 변수를 외부에 노출하지 
않고도 최솟값과 최댓값을 찾을 수 있습니다.
 ● find_min_max 함수가 호출된 이후에도 find_min과 find_max 함수를 
통해 min_value와 max_value 변수에 접근할 수 있습니다

```python
def find_min_max(numbers):
    # 내장 함수를 사용해 즉시 계산하고 클로저로 반환
    return lambda: min(numbers), lambda: max(numbers)

# 사용 예시
numbers = [10, 5, 8, 12, 3, 7]
find_min, find_max = find_min_max(numbers)

print("최솟값:", find_min())  # 3
print("최댓값:", find_max())  # 12
```

# [QUEST 03] 클로저를 활용한 문제 해결하기

```python
def counter(func):
    count = 0
    def wrapper(*args, **kwargs):
        nonlocal count
        count += 1
        # 먼저 원본 함수를 실행합니다.
        result = func(*args, **kwargs)
        # 그 다음에 실행 횟수를 출력합니다.
        print(f"{func.__name__} 실행횟수: {count}")
        return result
    return wrapper

@counter
def say_hello():
    print("Hello Aiffel!")

for i in range(5):
    say_hello()
```
