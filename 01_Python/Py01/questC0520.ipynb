# Q1-1 회문 판독기

```python
def is_palindrome(a):
    return a == a[::-1]

target1 = "아많다많다많다많아"
target2 = "아정말많다많다많아"
print(str(is_palindrome(target1)) + "\n" + str(is_palindrome(target2)))
```

# Q1-2 회문 분류기

```python
def is_palindrome(a):
    return a == a[::-1]

def palindrome_classifier(targets):
    o_list = [w for w in targets if is_palindrome(w)]
    x_list = [w for w in targets if not is_palindrome(w)]
    return {"회문O": o_list, "회문X": x_list}, len(o_list), len(o_list) / len(targets)

targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']
print(palindrome_classifier(targets))
```

# Q2-1 성적 데이터화

```python
scores = {
    "유원영": {"국어": 94, "영어": 91, "수학": 89},
    "설지민": {"국어": 88, "영어": 96, "수학": 90},
    "신윤아": {"국어": 86, "영어": 88, "수학": 77},
    "장윤아": {"국어": 98, "영어": 76, "수학": 92}
}
print(scores)
```

# Q2-2 성적 분석기

```python
def analyze_scores(student_name):
    student_data = scores[student_name]
    class_averages = {"국어": 91.5, "영어": 87.75, "수학": 87.0}
    
    print(f"[{student_name} 성적 분석]")
    total_score = 0
    for subject, score in student_data.items():
        total_score += score
        class_avg = class_averages[subject]
        print(f"- {subject} : {score}점 / 학급평균({class_avg:.2f}점) 대비 {score - class_avg:+.2f}점")
        student_avg = total_score / len(student_data)
    print(f"- 전과목 평균 : {student_avg:.2f}점 / 학급평균(88.75점) 대비 {student_avg - 88.75:+.2f}점")

    
analyze_scores("설지민")
```
