(입금내역과출금내역을출력하는deposit_history 와withdraw_history  메서드를추가하세요.)


```python
import random  # 랜덤한 숫자를 뽑기 위해 필요한 파이썬 내장 도구입니다.

# =============================================
# 퀘스트 1번: 은행계좌 만들기
# =============================================

class Account:
    # Q2. 클래스 변수: 모든 Account 인스턴스가 공유하는 '전체 계좌 수'를 저장합니다.
    account_count = 0  

    def __init__(self, owner, balance):  # Q1. 생성자: 계좌가 만들어질 때 자동으로 실행되어 초기 정보를 설정합니다.
        self.bank_name = "SC은행"
        self.owner = owner
        self.balance = balance
        # 계좌번호를 랜덤으로 생성하는 함수를 호출하여 저장합니다.
        self.account_number = self._generate_account_number()
        # Q7. 입금 횟수를 세기 위한 변수입니다.
        self._deposit_count = 0 
        # Q10. 입출금 내역을 순서대로 저장할 빈 리스트를 준비합니다.
        self._deposit_log = [] 
        self._withdraw_log = [] 
        # Q2. 계좌가 하나 새로 생길 때마다 전체 개수를 1씩 늘립니다.
        Account.account_count += 1 

    def _generate_account_number(self):
        """계좌번호 랜덤 생성 (111-11-111111 형태)"""
        part1 = str(random.randint(100, 999))
        part2 = str(random.randint(10, 99))
        part3 = str(random.randint(100000, 999999))
        return f"{part1}-{part2}-{part3}"

    @classmethod
    def get_account_num(cls):  # Q3. 클래스 메서드: 인스턴스 없이도 현재까지의 총 계좌 수를 출력합니다.
        print(f"현재 생성된 계좌 수: {cls.account_count}개")

    def deposit(self, amount):  # Q4. 입금 메서드: 계좌에 돈을 넣습니다.
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return
        self.balance += amount
        self._deposit_count += 1
        self._deposit_log.append(amount)  # Q10. 입금 내역을 기록합니다.
        print(f"{amount:,}원이 입금되었습니다. 잔고: {self.balance:,}원")

        # Q7. 입금 5회마다 잔액의 1%를 이자로 더해주는 로직입니다.
        if self._deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print(f"🎉 입금 5회 달성! 이자 {interest:,}원이 지급되었습니다. 잔고: {self.balance:,}원")

    def withdraw(self, amount):  # Q5. 출금 메서드: 잔고가 충분할 때만 돈을 뺍니다.
        if amount > self.balance:
            print("잔고가 부족합니다. 출금할 수 없습니다.")
            return
        self.balance -= amount
        self._withdraw_log.append(amount)  # Q10. 출금 내역을 기록합니다.
        print(f"{amount:,}원이 출금되었습니다. 잔고: {self.balance:,}원")

    def display_info(self):  # Q6. 계좌 정보를 예쁘게 출력합니다.
        print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, "
              f"계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

    def deposit_history(self):  # Q10. 리스트에 저장된 입금 내역을 순서대로 보여줍니다.
        print(f"[{self.owner}] 입금 내역:")
        if not self._deposit_log:
            print("  입금 내역이 없습니다.")
        for i, amount in enumerate(self._deposit_log, 1):
            print(f"  {i}회: +{amount:,}원")

    def withdraw_history(self):  # Q10. 리스트에 저장된 출금 내역을 순서대로 보여줍니다.
        print(f"[{self.owner}] 출금 내역:")
        if not self._withdraw_log:
            print("  출금 내역이 없습니다.")
        for i, amount in enumerate(self._withdraw_log, 1):
            print(f"  {i}회: -{amount:,}원")


# =============================================
# 테스트 코드
# =============================================

print("=" * 50)
print("퀘스트 1번: 은행계좌 만들기")
print("=" * 50)

# Q8. 3개 이상 인스턴스(객체) 생성 후 리스트에 담아 관리합니다.
acc1 = Account("김철수", 500000)
acc2 = Account("이영희", 1500000)
acc3 = Account("박민준", 2000000)
acc4 = Account("최수연", 800000)

accounts = [acc1, acc2, acc3, acc4]

# Q3. 만들어진 전체 계좌 개수를 확인합니다.
Account.get_account_num()

print()

# Q6. 모든 계좌의 정보를 출력합니다.
print("[계좌 정보]")
for acc in accounts:
    acc.display_info()

print()

# Q4/Q5 입출금 기능을 테스트합니다.
print("[입출금 테스트 - 김철수 계좌]")
acc1.deposit(100000)
acc1.deposit(50000)
acc1.deposit(200000)
acc1.withdraw(30000)
acc1.deposit(10000)
acc1.deposit(20000)   # 5번째 입금 → 이자 지급이 확인됩니다!

print()

# Q10. 기록된 입출금 내역을 확인합니다.
acc1.deposit_history()
acc1.withdraw_history()

print()

# Q9. 반복문을 사용해 잔고가 100만 원 이상인 'VIP 고객'만 골라서 정보를 출력합니다.
print("[잔고 100만원 이상 고객]")
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()
```

○  전투도중에Player가 사망하면이후전투를진행하지않고 게임오버 출력하고종료  

```python
import random  # 게임에서 사용할 랜덤한 숫자(데미지, 능력치 등)를 뽑기 위한 도구입니다.

# =============================================
# 퀘스트 2번: 간단한 자동사냥 RPG 만들기
# =============================================

# Q1. Character 클래스: 모든 게임 캐릭터(플레이어와 몬스터)가 갖는 공통적인 틀입니다.
class Character:
    def __init__(self, name, level, hp, attack, defense):
        # 캐릭터의 고유 정보를 저장합니다.
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        # 캐릭터의 체력이 0보다 크면 살아있다고 판단합니다.
        return self.hp > 0

    def take_damage(self, damage):
        # 받은 공격력에서 나의 방어력을 뺀 만큼 체력이 깎입니다.
        actual_damage = damage - self.defense
        if actual_damage > 0:
            self.hp -= actual_damage
        # 방어력이 데미지보다 크면 체력이 깎이지 않습니다.

    def attack_target(self, target):
        # 1부터 나의 공격력 사이의 랜덤한 수치만큼 상대방에게 데미지를 줍니다.
        damage = random.randint(1, self.attack)
        target.take_damage(damage)
        # 공격 상황을 화면에 출력합니다.
        print(f"{self.name}이(가) {target.name}에게 {damage}만큼 공격했다...!")
        print(f"{target.name}의 체력: {target.hp}")


# Q2. Player 클래스: 캐릭터의 기능을 이어받아(상속), 플레이어만의 고유 기능을 가집니다.
class Player(Character):
    def __init__(self, name):
        # 부모 클래스(Character)의 초기 설정을 그대로 가져옵니다.
        super().__init__(
            name=name,
            level=1,
            hp=100,
            attack=25,
            defense=5
        )
        self.exp = 0  # 플레이어는 경험치라는 특별한 속성을 가집니다.

    def gain_experience(self, amount):
        # 적을 쓰러뜨리고 얻은 경험치를 내 정보에 더합니다.
        self.exp += amount
        print(f"{self.name}이(가) 경험치 {amount}를 획득했습니다! (현재 경험치: {self.exp})")

    def level_up(self):
        # 경험치가 50 이상 모일 때마다 레벨과 능력치가 자동으로 올라갑니다.
        while self.exp >= 50:
            self.exp -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            # 레벨업 축하 메시지와 변화된 능력치를 보여줍니다.
            print(f"🎉 레벨 업! {self.name}의 레벨이 {self.level}이 되었습니다! "
                  f"(공격력: {self.attack}, 방어력: {self.defense})")


# Q2. Monster 클래스: 캐릭터 기능을 이어받고, 몬스터만의 능력치 생성 공식을 가집니다.
class Monster(Character):
    def __init__(self, name, level):
        # 몬스터는 레벨이 높을수록 능력치도 비례해서 강해집니다.
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(
            name=name,
            level=level,
            hp=hp,
            attack=attack,
            defense=defense
        )


# Q3. battle 함수: 플레이어와 몬스터가 체력이 다 할 때까지 주고받는 전투 로직입니다.
def battle(player, monster):
    # 전투 시작을 알리는 경계선을 출력합니다.
    print(f"\n{'='*40}")
    print(f"{monster.name} 과의 전투를 시작합니다.")
    print(f"{'='*40}")

    # 두 캐릭터 중 한 명의 체력이 0이 될 때까지 공격을 반복합니다.
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)   # 플레이어가 먼저 공격합니다.
        if not monster.is_alive():      # 몬스터가 죽었는지 바로 확인합니다.
            break
        monster.attack_target(player)   # 살아있다면 몬스터가 반격합니다.

    # 전투 결과(승리/패배)를 확인하고 처리합니다.
    if player.is_alive():
        exp_gain = monster.level * 20
        player.gain_experience(exp_gain) # 승리 보상으로 경험치를 받습니다.
        player.level_up()               # 레벨업 조건을 체크합니다.
        print("전투 승리!")
        print(f"{monster.name} 을(를) 이겼다!")
    else:
        print("전투 패배..")


# Q4. main 함수: 게임 전체의 흐름을 조종하는 시작점입니다.
def main():
    # 사냥할 몬스터들의 이름과 레벨 리스트입니다.
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

    # 사용자에게 이름을 입력받아 플레이어를 생성합니다.
    player_name = input("플레이어 이름을 입력하세요: ")
    player = Player(player_name)
    print(f"\n{player.name} 님, 게임을 시작합니다! "
          f"(체력: {player.hp}, 공격력: {player.attack}, 방어력: {player.defense})\n")

    # 몬스터 딕셔너리에 있는 몬스터들을 하나씩 불러와서 전투를 진행합니다.
    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)
        battle(player, monster)

        # 만약 전투에서 져서 살아있지 않다면, 남은 전투를 멈추고 게임을 종료합니다.
        if not player.is_alive():
            print("\n게임오버")
            return

    # 모든 전투를 통과하면 최종 메시지를 출력합니다.
    print(f"\n🏆 모든 몬스터를 물리쳤습니다! 최종 레벨: {player.level}")


# 이 파일을 직접 실행했을 때 main 함수를 호출하여 게임을 시작합니다.
if __name__ == "__main__":
    main()
```

# **회고**

```python
상속(Inheritance)의 효율성: super().__init__를 활용해 부모 클래스의 속성을 체계적으로 물려받는 과정을 실습하며, 중복 코드를 제거하고 관리 효율을 높이는 엔지니어링 설계를 경험했습니다.

로직의 캡슐화와 방어적 프로그래밍: take_damage 메서드에서 max(0, damage - defense)를 활용하여 방어 로직의 무결성을 지키고, is_alive를 통해 상태를 제어하는 과정을 통해 안정적인 시스템 구축의 중요성을 배웠습니다.

게임 루프(Game Loop) 설계: while문을 활용한 전투 로직과 딕셔너리를 활용한 몬스터 관리 등, 동적인 데이터를 순차적으로 처리하는 알고리즘적 사고를 길렀습니다.

반성

의미 있는 주석의 가치: 이번 과제에서 코드 로직을 건드리지 않고 주석만으로 가독성을 확보하는 과정을 거치며, '좋은 코드는 읽기 쉬운 코드'라는 원칙을 다시금 되새겼습니다.

향후 과제: 이제 기본적인 상속을 넘어, '다형성(Polymorphism)'을 활용해 더 다양한 몬스터 타입이나 스킬 시스템을 확장해 보는 과정이 필요합니다.
```
