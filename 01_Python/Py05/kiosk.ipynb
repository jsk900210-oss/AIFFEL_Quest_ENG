# 키오스크 만들기

### 노드10. 키오스크 만들기

![](./imege/0.png)  

```python
class Kiosk:
    def __init__(self):
        # 1. 초기 데이터 설정
        self.menu = ['아메리카노', '라떼', '모카', '유자차', '녹차', '초코라떼']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []
        self.order_price = []

    # 메뉴 출력 메서드
    def menu_print(self):
        print("\n--- 메뉴판 ---")
        for i in range(len(self.menu)):
            print(f"{i + 1}. {self.menu[i]} : {self.price[i]}원")

    # 주문 및 추가 주문 메서드
    def menu_select(self):
        while True:
            self.menu_print()
            try:
                n = int(input("\n음료 번호를 입력하세요 (종료: 0) : "))
                if n == 0:
                    print("주문을 종료합니다.")
                    break
                if 1 <= n <= len(self.menu):
                    t = int(input("HOT(1) or ICE(2) : "))
                    self.temp = "HOT" if t == 1 else "ICE"
                    self.order_price.append(self.price[n-1])
                    self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                    print(f"{self.order_menu[-1]}가 추가되었습니다.")
                else:
                    print("없는 메뉴입니다. 다시 주문해 주세요.")
            except ValueError:
                print("숫자만 입력해 주세요.")

    # 지불 메서드
    def pay(self):
        total = sum(self.order_price)
        print(f"\n총 결제 금액은 {total}원입니다.")
        while True:
            method = input("결제 수단을 선택하세요 (1: 현금, 2: 카드): ")
            if method in ['1', 'cash']:
                print("직원을 호출하겠습니다.")
                break
            elif method in ['2', 'card']:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("다시 결제를 시도해 주세요.")

    # 주문서 출력 메서드
    def table(self):
        print('\n⟝' + '-' * 30 + '⟞')
        for i in range(len(self.order_menu)):
            print(f"| {self.order_menu[i]} : {self.order_price[i]}원")
        print('-' * 32)
        print(f"| 합계 금액 : {sum(self.order_price)}원")
        print('⟝' + '-' * 30 + '⟞\n')

# [실행부]
a = Kiosk()
a.menu_select()  # 주문 시작
a.table()        # 영수증 확인
a.pay()          # 결제 진행
```
